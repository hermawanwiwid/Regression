# REGRESI NON LINEAR

```
url2 = 'https://raw.githubusercontent.com/hermawanwiwid/Regression/master/kuadratik.csv'

data = pd.read_csv(url2)

data.head()
```

```
import matplotlib.pyplot as plt
cdf = data[['x','ydata']]
plt.scatter(cdf.x, cdf.ydata,  color='blue')
plt.xlabel("x")
plt.ylabel("ydata")
plt.show()
```

```
X = cdf.x
Y = cdf.ydata
X_mean = np.mean(X)
Y_mean = np.mean(Y)

num = 0
den = 0
for i in range(len(X)):
    num += (X[i] - X_mean)*(Y[i] - Y_mean)
    den += (X[i] - X_mean)**2
m = num / den
c = Y_mean - m*X_mean

print (m, c)
```

```
sumx = np.sum(X)
sumy = np.sum(Y)
sumx2 = np.sum(X**2)
sumxy = X @ Y
m = len(X)
a = np.array([[m, sumx], [sumx,sumx2]])
b = np.array([sumy,sumxy])
xBar = np.linalg.inv(a) @ b.T

print(xBar)
# print(np.linalg.inv(a))
```

Mencoba menggunakan linear regression


```
plt.scatter(cdf.x, cdf.ydata,  color='blue')
plt.plot(X,xBar[1]*X+xBar[0], color ='r')
plt.xlabel("x")
plt.ylabel("ydata")
plt.show()

X = cdf.x
Y = cdf.ydata
```

Hampiran polinomial

```
sumx = np.sum(X)
sumy = np.sum(Y)
sumx2 = np.sum(X**2)
sumx3 = np.sum(X**3)
sumx4 = np.sum(X**4)
sumxy = X @ Y
sumyx2 = Y @ X**2
m = len(cdf.x)
a = np.array([[m,sumx,sumx2],[sumx,sumx2,sumx3],[sumx2,sumx3,sumx4]])
b = np.array([sumy,sumxy,sumyx2])
xBar = np.linalg.inv(a) @ b.T

print(xBar)
# print(np.linalg.inv(a))
```

# EKSPONENSIAL

```
url3 = 'https://raw.githubusercontent.com/hermawanwiwid/Regression/master/eksponensial.csv'

data = pd.read_csv(url3)

data.head()
```

```
import matplotlib.pyplot as plt
cdf = data[['x','ydata']]
plt.scatter(cdf.x, cdf.ydata,  color='blue')
plt.xlabel("x")
plt.ylabel("ydata")
plt.show()
```

```
lny = np.log(abs(cdf.ydata))
# print(cdf.ydata)
# lny
```

```

X = cdf.x
Y = cdf.ydata
```

```
sumx = np.sum(X)
sumy = np.sum(lny)
sumx2 = np.sum(X**2)
sumxy = X @ lny
xy = np.sum(X*lny)
m = len(X)
a = np.array([[m, sumx], [sumx,sumx2]])
b = np.array([sumy,xy])
xBar = np.linalg.inv(a) @ b
lnxBar = np.log(abs(xBar))
alfa = np.exp(xBar[1])
# print(xBar)
def ln_y(x):
  return x[1]*np.exp(alfa*x)

# print(np.linalg.inv(a))
```

```
import math

plt.scatter(cdf.x, cdf.ydata,  color='magenta')
# plt.plot(X,xBar[1]*X+xBar[0], color ='b')
# plt.plot(X,lnxBar[0]*math.e**(xBar[1]*X), color ='b')
plt.plot(X,ln_y(X), color ='b')
plt.plot()
plt.xlabel("x")
plt.ylabel("ydata")
plt.show()
```
